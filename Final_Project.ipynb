```python
df = df.dropna();
```

Our data is sorted by date but in a reverse way. I turn that into earliest to latest

```python
df = df.sort_values("GAME_DATE_EST")
df = df.reset_index(drop=True)
df
```

```python
def next_match_score(df):
    df["target"]= df["HOME_TEAM_WINS"].shift(-1)
    return df
```

We add next match score for each home team to test wether we are accurate or not.

```python
df = df.groupby("HOME_TEAM_ID").apply(next_match_score)
```

```python
df= df.dropna()
df
```

We dropped the NULL values because there is no match after last match. The last value comes as NULL

```python
df["HOME_TEAM_WINS"].value_counts()
```

We can see that if the match is home there is 60% probability of winning for the home team.

```python
from sklearn.model_selection import TimeSeriesSplit 
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

R = RidgeClassifier(alpha = 1)
split = TimeSeriesSplit(n_splits = 10)
sfs = SequentialFeatureSelector(R, n_features_to_select = 10, direction="forward", cv = split )
```

```python
df.columns
```

```python
removal_columns = ['GAME_DATE_EST', 'GAME_ID', 'GAME_STATUS_TEXT', 'HOME_TEAM_ID',
                   'VISITOR_TEAM_ID', 'SEASON', 'TEAM_ID_home', 'TEAM_ID_away','target']
```

```python
usable_columns = df.columns[~df.columns.isin(removal_columns)]
```

```python
usable_columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[usable_columns] = scaler.fit_transform(df[usable_columns])
```

```python
df
```

```python
sfs.fit(df[usable_columns], df["target"])
```

```python
predictors = list(usable_columns[sfs.get_support()])
```

```python
predictors
```

```python
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []
    seasons = sorted(data["SEASON"].unique())
    
    for i in range(start, len(seasons), step):
        season = seasons[i]
        train  = data[data["SEASON"] < season]
        test   = data[data["SEASON"] == season]
        
        model.fit(train[predictors], train["target"])
        
        preds = model.predict(test[predictors])
        
        preds = pd.Series(preds, index=test.index)
        
        combined = pd.concat([test["target"], preds], axis=1)
        
        combined.columns = ["actual", "predicted"]
        
        all_predictions.append(combined)
        
    return  pd.concat(all_predictions)
```

This function is to take the season we want and train the algorithm by previous season and make the guess.

```python
predictions = backtest(df, R, predictors)
predictions
```

```python
predictions.dropna();
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(predictions["actual"], predictions["predicted"])
```

We can guess in a %59 rate for winning team
